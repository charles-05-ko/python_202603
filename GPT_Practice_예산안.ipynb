```python
text = "커피 빵 빵 커피 라면 커피"
words = text.split()
words
```

```python
counts = {}
for w in words:
    counts[w] = counts.get(w, 0) + 1
    counts
```

```python
top_word = max(counts, key=counts.get)
print("최다 단어", top_word, "횟수", counts[top_word])
```

```python
answers = ["A", "C", "B", "D"]
my = ["A", "B", "B", "D"]
```

```python
score = 0
results = []
for a, m in zip(answers, my):
    ok = (a==m)
    results.append(ok)
    if ok:
        score += 1

results, score
```

```python
print("정답 여부:", results)
print("점수:", score, "/", len(answers))
```
