## 5. 딕셔너리 관련 함수

### keys() : key 리스트 만들기

```python
dic = {'name':'주철승',  'phone':'010-1111-2222', 'brith': '1111'}
dic.keys()
```

```python
list(dic.keys())
```

```python
for key in dic.keys():
    print(key)
```

### values() : Value 리스트 만들기

```python
dic.values()
```

```python
for value in dic.values():
    print(value)
```

### items() : key, value 쌍 얻기

```python
dic.items()
```

### clear() : key, value 쌍 모두 지우기

```python
dic
```

```python
dic.clear()
dic
```

### grt() : key로 value 얻기

```python
a = {'name':'주철승',  'phone':'010-1111-2222', 'brith': '1111'}
a
```

```python
a['name']
```

```python
a.get('name')
```

```python
print(a.get('age'))
```

```python
a['age']
```

```python
a.get('age', 20)
```

```python
a['age'] = 25
a
```

```python
a.get('age', 20)
```
